```python
import csv
class EmployeeManager:
    def __init__(self,ID,Name,Position,Salary,Email):
        self.ID=ID
        self.Name=Name
        self.Position=Position
        self.Salary=Salary
        self.Email=Email

        def display_info(self):
            print(f"ID: {self.ID}")
            print(f"Name: {self.Name}")
            print(f"Position: {self.Position}")
            print(f"Salary: {self.Salary}")
            print(f"Email: {self.Email}")
            print("-" * 30) # will prin - 30 times

```

```python
employees ={}
while True:
    emp_ID=input("Enter Emloyee ID: ")
    emp_Name=input("Enter Name: ")
    emp_Position=input("Enter Position: ")
    emp_Salary=input("Enter Salary: ")
    emp_Email=input("Enter Email: ")
    emp=EmployeeManager(emp_ID,emp_Name,emp_Position,emp_Salary,emp_Email)
    employees[emp_ID]=emp
    choice = input("Add another employee? (y/n): ")
    if choice.lower() =='n':
        break

# csv header
fieldnames = ['ID','Name', 'Position', 'Salary', 'Email']

# csv data
rows=[]
for emp in employees.values():
    rows.append({
    'ID':emp.ID,   
    'Name':emp.Name,
    'Position':emp.Position,
    'Salary':emp.Salary,
    'Email':emp.Email
    }
    )

with open('employees.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)
   
```
